```python
import networkx as nx
import matplotlib.pyplot as plt

#Create an empty graph with no nodes and no edges.
G = nx.DiGraph()
```

```python
#Add Nodes

G.add_node('1', level ='System', name='Car')
G.add_node('2', level ='Subsystem', name='Interior')
G.add_node('3', level ='Subsystem', name='Body')
G.add_node('4', level='Subsystem', name='Engine')
G.add_node('5', level='Subsystem', name='Steering System')
G.add_node('6', level='Part', name='Batteries')
G.add_node('7', level='SubSystem', name='Chassie')
G.add_node('8', level='Part', name='Tires')
G.add_node('9', level='Person', name='Manufacturer')
G.add_node('10', level='Person', name='User')
G.add_node('11', level='Person', name='Seller')
G.add_node('12', level='Subsystem', name='Gearbox')
G.add_node('14', level='Subsystem', name='Automatic')
G.add_node('15', levell='Subsystem', name='Manual')
G.add_node('16', level='Subsystem', name='Otto Engine')
G.add_node('17', level='Subsystem', name='Hybrid Engine')
G.add_node('18', level='Subsystem', name='Electric Engine')
G.add_node('19', level='Part', name='Door')
G.add_node('20', level='Part', name='Window')
```

```python
#Add Edges
G.add_edges_from([('1', '2'), ('1', '3'), ('1', '4'),('1','5'),('1','7'),('4','12'),('12','14'),('12','15'),('4','16'),('4','17'),('4','18')],
                 type='SUBSYSTEM_OF')

G.add_edges_from([('1', '6'), ('8', '5'), ('3', '19'),('3','20')],
                 type='PART_OF')

G.add_edge('1', '10', type='USES')

G.add_edge('11', '1', type='CONFIGURES')

G.add_edge('11', '10', type='SELLS_TO')

G.add_edge('11', '9', type='ORDERS_BY')

G.add_edge('9', '1', type='MANUFACTURES')

G.add_edge('9', '11', type='DELIVERS_TO')

G.add_edge('10', '11', type='BUYS_BY')



```

```python
pos = nx.spring_layout(G, k=0.8, iterations=20)
# k controls the distance between the nodes and varies between 0 and 1
# iterations is the number of times simulated annealing is run
# default k=0.1 and iterations=50

#pos = nx.circular_layout(G, scale=1, center=None)

#Size of the plot
plt.figure(figsize=(15,15))

nx.draw_networkx(G, pos, node_size = 500)
node_labels = nx.get_node_attributes(G,'name')
nx.draw_networkx_labels(G, pos, labels = node_labels)

edge_labels = nx.get_edge_attributes(G,'type')
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_labels)

plt.show()
```
