```python
import networkx as nx
import matplotlib.pyplot as plt

#Create an empty graph with no nodes and no edges.
G = nx.DiGraph()
```

```python
#add nodes

G.add_node('1', Name="Airport")
G.add_node('2', Name="Aircaft DLH44C")
G.add_node('3', Name="Passenger")
G.add_node('4', Name="Ticket")
G.add_node('5', Name="BER")
G.add_node('6', Name="LHR")
G.add_node('7', Name="Airline")
G.add_node('8', Name="Lufthansa")
G.add_node('9', Name="Crew")
G.add_node('10', Name="Pilot")
G.add_node('11', Name="Attendant")
G.add_node('12', Name="AircraftSize 303")
G.add_node('13', Name="AircraftType A350")
G.add_node('14', Name="Flight Memgraph")
G.add_node('15', Name="DepartureTime")
G.add_node('16', Name="ArrivalTime")
G.add_node('17', Name="FirstOfficer")
G.add_node('18', Name="Lufthansa Technik")
G.add_node('19', Name="Route")
G.add_node('20', Name="Supplier")
G.add_node('21', Name="Catering")
```

```python
#add edges

G.add_edge('14', '2', type='ASSIGNED_AIRCRAFT')
G.add_edge('2', '12', type='AIRCRAFT_SIZE')
G.add_edge('2', '13', type='AIRCRAFT_TYPE')
G.add_edge('14', '5', type='DEPARTURE_AIRPORT')
G.add_edge('14', '6', type='ARRIVAL_AIRPORT')
G.add_edge('5', '1', type='is a')
G.add_edge('6', '1', type='is a')
G.add_edge('14', '8', type='OPERATED_BY')
G.add_edge('9', '14', type='IS_ASSIGNED_TO')
G.add_edge('10', '9', type='IS_ASSIGNED_TO')
G.add_edge('17', '9', type='IS_ASSIGNED_TO')
G.add_edge('11', '9', type='IS_ASSIGNED_TO')
G.add_edge('2', '18', type='IS_MAINTAINED_BY')
G.add_edge('2', '8', type='IS_OPERATED_BY')
G.add_edge('14', '15', type='DEPARTURE')
G.add_edge('14', '16', type='ARRIVAL')
G.add_edge('3', '4', type='BOOKED')
G.add_edge('4', '8', type='BOOKED_BY')
G.add_edge('8', '7', type='is a')
G.add_edge('4', '19', type='HAS_PLANNED_ROUTE')
G.add_edge('20', '21', type='DELIVERS')
G.add_edge('21', '2', type='DELIVERED_TO')
```

```python
pos = nx.spring_layout(G, k=0.8, iterations=20)
# k controls the distance between the nodes and varies between 0 and 1
# iterations is the number of times simulated annealing is run
# default k=0.1 and iterations=50

#pos = nx.circular_layout(G, scale=1, center=None)

#Size of the plot
plt.figure(figsize=(15,15))

nx.draw_networkx(G, pos, node_size = 500)
node_labels = nx.get_node_attributes(G,'Name')
nx.draw_networkx_labels(G, pos, labels = node_labels)

edge_labels = nx.get_edge_attributes(G,'type')
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_labels)

plt.show()
```
